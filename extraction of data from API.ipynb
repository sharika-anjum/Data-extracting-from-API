```python
url = "https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page="
page_list = []
# For each page, we access its information through its API
for channel in range(1,11):
    channel=url+str(channel)
    JC=requests.get(url).json()
    for i in JC['results']:
        lists=[]
        lists.append(i["id"])
        lists.append(i["title"])
        lists.append(i["overview"])
        lists.append(i["release_date"])
        lists.append(i["popularity"]) 
        lists.append(i["vote_average"])
        page_list.append(lists) 
    
    
page_list
```

```python
dataset = pd.DataFrame(page_list)
dataset
```

```python
dataset.columns = ['ID','Title', 'Overview', 'Release date', 'Popularity', 'Average Vote']
```
